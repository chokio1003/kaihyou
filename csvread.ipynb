```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
import numpy
import os
import pprint
from scipy.interpolate import RectBivariateSpline

#最小値0、最大値1にする数量正規化メソッド
def normalization(x):
    x_min = x.min()
    x_max = x.max()
    x_norm = (x - x_min) / (x_max - x_min)
    return x_norm   #正規化した配列を返す
```

```python
import pandas as pd

df1 = pd.read_csv("csv/201603270300gakushu.csv")
df1 = df1[["1","2","3","17"]]
df2 = pd.read_csv("kekka.csv")
df2 = df2[["1","0"]]                  
```

```python
datar = df1["3"]
datag = df1["2"]
datab = df1["1"]
data1 = df2["1"]
data0 = df2["0"]
```

```python
r=np.arange(300000) 
r=datar
R = np.empty([500, 600])
h = 0
for i in range(600):
    for j in range(500):
        R[j][i] = r[h]
        h = h+1
```

```python
g=np.arange(300000) 
g=datag
G = np.empty([500, 600])
h = 0
for i in range(600):
    for j in range(500):
        G[j][i] = g[h]
        h = h+1
```

```python
b=np.arange(300000) 
b=datab
B = np.empty([500, 600])
h = 0
for i in range(600):
    for j in range(500):
        B[j][i] = b[h]
        h = h+1
```

```python
s0=np.arange(300000)
s1=np.arange(300000) 
s0=data0
s1=data1
M = np.empty([500, 600])
h = 0
for i in range(600):
    for j in range(500):
        if s1[h] > 0.6:
            M[j][i] = 1
        else:
            M[j][i] = 0
        h = h+1
```

```python
plt.imshow(M)
```

```python
R = normalization(R)
G = normalization(G)
B = normalization(B)
for i in range(600):
    for j in range(500):
        if M[j][i]==1:
            R[j][i]=1
            G[j][i]=0
            B[j][i]=0
rgb = np.dstack((R, G, B))
rgb = np.power(rgb, 1.0)
plt.imshow(rgb);
```
